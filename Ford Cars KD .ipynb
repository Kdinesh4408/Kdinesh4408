***CLEANING RATE :***

```python
def handle_EngineSize(value):
    if(value=='NEW' or value=='_'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
ad['EngineSize']=ad['EngineSize'].apply(handle_EngineSize)
ad['EngineSize'].head()
```
